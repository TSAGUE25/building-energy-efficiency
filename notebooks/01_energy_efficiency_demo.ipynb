## 1. Imports et configuration

Chargement des librairies necessaires.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
np.random.seed(42)
print('Imports OK')
```

## 2. Generation des donnees simulees

Creation d'un dataset de batiments avec consommation energetique.


```python
N = 500
df = pd.DataFrame({
    'surface_m2':      np.random.uniform(100, 5000, N),
    'annee_construction': np.random.randint(1960, 2023, N),
    'nb_etages':        np.random.randint(1, 20, N),
    'type_batiment':    np.random.choice(['Bureau','Logement','Commerce'], N),
    'conso_kwh':        np.abs(np.random.normal(250000, 80000, N)),
})
df['iep_kwh_m2'] = df['conso_kwh'] / df['surface_m2']
print(f'Dataset: {len(df)} batiments')
print(df.describe().round(1))
```

## 3. Segmentation KMeans

Identification de 3 profils energetiques par clustering.


```python
features = ['iep_kwh_m2', 'surface_m2', 'annee_construction']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['segment'] = kmeans.fit_predict(X_scaled)

print('Segments identifies:')
print(df.groupby('segment')['iep_kwh_m2'].agg(['mean','count']).round(1))
```

## 4. Visualisation des segments

Representation graphique des 3 profils de consommation.


```python
fig, ax = plt.subplots(figsize=(9, 5))
colors = ['#2ecc71', '#e67e22', '#e74c3c']
labels = ['Performant', 'Moyen', 'Energie++']
for seg in [0, 1, 2]:
    mask = df['segment'] == seg
    ax.scatter(df[mask]['surface_m2'], df[mask]['iep_kwh_m2'],
               c=colors[seg], alpha=0.6, s=20, label=labels[seg])
ax.set_xlabel('Surface (m2)')
ax.set_ylabel('IEP (kWh/m2)')
ax.set_title('Segmentation energetique des batiments')
ax.legend()
plt.tight_layout()
plt.savefig('../figures/segments_energetiques.png', dpi=120)
plt.show()
print('Graphique sauvegarde dans figures/')
```
